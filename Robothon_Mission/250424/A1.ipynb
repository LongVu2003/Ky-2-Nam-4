```python
 from pyLeanbot import *

deg90 = 68
l = 170     # length
w = 140     # width

Leanbot.begin()

def MoveO(mm):
    LbMotion.runLR(+2000, +2000)
    LbMotion.waitDistanceMm(mm)
    LbMotion.stopAndWait()

def MoveB(mm):
    LbMotion.runLR(-2000, -2000)
    LbMotion.waitDistanceMm(mm)
    LbMotion.stopAndWait()

def TurnR(deg):
    LbMotion.runLR(+1000, -1000)
    LbMotion.waitRotationDeg(deg)
    LbMotion.stopAndWait()

def TurnL(deg):
    LbMotion.runLR(-1000, +1000)
    LbMotion.waitRotationDeg(deg)
    LbMotion.stopAndWait()

# --- Start ---
LbMission.beginDigitalTwin("msNumberGrid_A1")

# --- 1. Grid 3 --> Grid 4 ---
# start at Grid 11
MoveB( 1 * w )			# stop at Grid 10
TurnL( deg90 )
MoveO( 1 * l )			# stop at Grid 3
LbGripper.close()

MoveB( 1 * l )			# stop at Grid 10
TurnR( deg90 )
MoveO( 3 * l - 25 )		# stop at Grid 13
TurnL( deg90 )
MoveO( 1 * l )			# stop at Grid 6
TurnL( deg90 )
MoveO( 2 * l - 50 )		# stop at Grid 4
LbGripper.open()

# --- 2. Grid 20 --> Grid 8 ---
# start at Grid 4
MoveB( 2 * l - 40 )		# stop at Grid 6
TurnL( deg90 -1)
MoveO( 1 * l )			# stop at Grid 13
TurnL( deg90 )
MoveO( 1 * w -20)			# stop at Grid 14
TurnR( deg90 +1)
MoveO( 1 * l )			# stop at Grid 21
TurnR( deg90 )
MoveO( 1 * w -10  )			# stop at Grid 20
LbGripper.close()

MoveB( 1 * w )			# stop at Grid 21
TurnR( deg90 )
MoveO( 1 * l )        		# stop at Grid 14
TurnL( deg90 )
MoveO( 6 * w )		 	# stop at Grid 9
TurnR( deg90 + 1 )
MoveO( 1 * l - 30 )         	# stop at Grid 2
TurnL( deg90 )
MoveO( 1 * w )			# stop at Grid 1
TurnL( deg90 )
MoveO( 1 * l -15 )        		# stop at Grid 8
LbGripper.open()

# --- Finish ---
LbMission.end()
```
